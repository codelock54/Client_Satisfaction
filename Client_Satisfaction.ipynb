```r
# Metodo del Codo
library(factoextra) 
fviz_nbclust(Data, hcut, method = "wss") + 
  ggtitle("Método de Codo para Determinar el Número de Clusters")
```

```r
set.seed(123)
k <- 4
kmeans_codo <- kmeans(Data, centers = k, nstart = 50)

kmeans_codo
```

```r
fviz_cluster(kmeans_codo, data = Data, 
             geom = "point", ellipse.type = "norm", 
             ggtheme = theme_minimal()) +
  ggtitle("Visualización de Clusters de Satisfacción del Cliente")
```

```r
# Metodo Silueta
fviz_nbclust(Data, kmeans, method = "silhouette") + 
  ggtitle("Método de Codo para Determinar el Número de Clusters")
```

```r
set.seed(123)
k <- 2
kmeans_silueta <- kmeans(Data, centers = k, nstart = 50)

kmeans_silueta
```

```r
fviz_cluster(kmeans_silueta, data = Data, 
             geom = "point",
             ggtheme = theme_minimal()) +
  ggtitle("Visualización de Clusters de Satisfacción del Cliente")
```

```r
Data$cluster <- kmeans_silueta$cluster

Data$cluster
```

```r
resumen_clusters <- Data %>%
  group_by(cluster) %>%
  summarise(across(everything(), list(media = \(x) mean(x, na.rm = TRUE), desviacion = \(x) sd(x, na.rm = TRUE))))
  
resumen_clusters
```

```r
ggplot(Data, aes(x = as.factor(cluster), y = Satisfaccion, fill = as.factor(cluster))) +
  geom_boxplot() +
  labs(title = "Distribución de Satisfacción por Cluster", x = "Cluster", y = "Satisfacción")
```

```r
anova_satisfaccion <- aov(Satisfaccion ~ cluster, data = Data)
summary(anova_satisfaccion)
```

```r
y <- Data$Satisfaccion
y
```

```r
x <- Data[, -((ncol(Data) - 1):ncol(Data))]
x
```

```r

pca_result <- prcomp(x, center = TRUE, scale. = TRUE)

summary(pca_result)

```

```r
factors <- as.factor(Data$cluster)
```

```r
library(ggplot2)
library(factoextra)

fviz_pca_ind(pca_result,
             geom.ind = "point",
             col.ind = factors,
             palette = "RdYlBu",
             addEllipses = TRUE, 
             legend.title = "Cluster") +
  ggtitle("PCA - Distribución de Clusters") +
  theme_minimal()
```

```r
x <- data.frame(principal_component = pca_result$x[, 1])
x
```

```r
pca_data <- data.frame(x, y)
pca_data
```

```r
fviz_nbclust(pca_data, hcut, method = "wss",k.max = 10)+
  labs(title= "Número optimo de cluster") + 
  xlab("Valor de k ") +
  ylab("Suma de cuadrados entre grupos")
```

```r
k <- 4
kmeans_pca <- kmeans(pca_data, centers = k, nstart = 50)

kmeans_pca
```

```r
pca_data$cluster <- kmeans_pca$cluster
pca_data
```
