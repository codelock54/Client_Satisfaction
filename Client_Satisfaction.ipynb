```r
rownames(Data) <- Data[[1]]

Data <- Data[, -1]

head(Data)
```

```r
library(GGally)

ggpairs(Data, 
        title = "Correlaciones con Distribuciones", 
        upper = list(continuous = wrap("cor", method = "pearson")), 
        lower = list(continuous = "smooth"),
        diag = list(continuous = "densityDiag"))
```

```r
corr_matrix <- cor(Data, use = "pairwise.complete.obs")
corr_matrix
```

```r
heatmap(corr_matrix, 
        main = "Matriz de Correlación", 
        col = heat.colors(256), 
        scale = "column", 
        margins = c(10, 10))
```
