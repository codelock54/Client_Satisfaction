## Analisis de Clustering con PCA

Este análisis tiene como objetivo realizar un **clustering** en los datos de satisfacción del cliente, utilizando el **Análisis de Componentes Principales (PCA)** para reducir la dimensionalidad antes de aplicar un algoritmo de clustering.


```r
library(factoextra)
pca_data <- prcomp(x, scale = TRUE)
fviz_eig(pca_data) 
```

```r
pca_data
```

```r

explained_variance <- (pca_data$sdev^2) / sum(pca_data$sdev^2) 
explained_variance
```

```r
cumulative_variance <- cumsum(explained_variance)
cumulative_variance
```

```r
num_components <- which(cumulative_variance >= 0.8)[1]
pca_scores <- as.data.frame(pca_data$x[, 1:num_components])
head(pca_scores,10)
```

```r
fviz_eig(pca_data, addlabels = TRUE, ylim = c(0, 100)) +
  ggtitle("Scree Plot - Varianza Explicada por Componentes Principales") +
  theme_minimal()
```

##  Análisis Cluster con Kmeans

```r
library(factoextra) 
fviz_nbclust(pca_scores, kmeans, method = "wss") + 
  ggtitle("Método de Codo para Determinar el Número de Clusters")
```

```r
fviz_nbclust(pca_scores, kmeans, method = "silhouette") +
  ggtitle("Método de la Silueta para Determinar el Número Óptimo de Clusters") +
  theme_minimal()
```

```r
set.seed(123)  
kmedias_codo <- kmeans(pca_scores, centers = 4)
kmedias_codo
```

```r
set.seed(123)  
kmedias_silhouette <- kmeans(pca_scores, centers = 2)
kmedias_silhouette
```

```r
library(factoextra)

fviz_cluster(kmedias_codo, data = pca_scores, geom = "point", 
             ellipse.type = "convex", 
             ggtheme = theme_minimal(),
             palette = "jco") +
  ggtitle("Clustering basado en Componentes Principales Metodo Codo") +
  labs(color = "Cluster")
```

```r
fviz_cluster(kmedias_silhouette, data = pca_scores, geom = "point", 
             ellipse.type = "convex", 
             ggtheme = theme_minimal(),
             palette = "jco") +
  ggtitle("Clustering basado en Componentes Principales Metodo silueta") +
  labs(color = "Cluster")
```

```r
Data_clusters_codo <- Data_scaled %>%
  mutate(cluster = as.factor(kmedias_codo$cluster))

```

```r
# Calcular la media de cada variable por cluster
resumen_clusters <- Data_clusters_codo %>%
  group_by(cluster) %>%
  summarise(across(everything(), list(media = ~mean(.x, na.rm = TRUE))))

resumen_clusters
```

```r
Data_clusters_silueta <- Data_scaled %>%
  mutate(cluster = as.factor(kmedias_silhouette$cluster))
```

```r
# Calcular la media de cada variable por cluster
resumen_clusters_silueta <- Data_clusters_silueta %>%
  group_by(cluster) %>%
  summarise(across(everything(), list(media = ~mean(.x, na.rm = TRUE))))

resumen_clusters_silueta
```
